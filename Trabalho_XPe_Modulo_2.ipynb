```python
import pandas as pd
```

```python
dados_pessoais = '/content/dados_pessoais.csv'
dados_fisicos = '/content/dados_caracteristicas_fisicas.csv'
dados_estados = '/content/estado_regiao.csv'
```

```python
df_pessoais = pd.read_csv(dados_pessoais, sep=';', encoding='latin1')
df_fisicos = pd.read_csv(dados_fisicos, sep=',', encoding='latin1')
df_estados = pd.read_csv(dados_estados, sep=';', encoding='latin1')

df_pessoais['cod_cliente'] = df_pessoais['cod_cliente'].astype(str).str.strip()
df_pessoais['cod_estado'] = df_pessoais['cod_estado'].astype(str).str.strip()
df_fisicos['ID Cliente'] = df_fisicos['ID Cliente'].astype(str).str.strip()
df_estados['id_estado'] = df_estados['id_estado'].astype(str).str.strip()
```

```python
df_merge_1 = pd.merge(df_pessoais, df_fisicos, left_on='cod_cliente', right_on='ID Cliente')
df_dados = pd.merge(df_merge_1, df_estados, left_on='cod_estado', right_on='id_estado')

# Remove a coluna 'Número de Cartões de Crédito'
df_dados.drop('Número de Cartões de Crédito', axis=1, inplace=True)

# realiza o filtro para aplicar a Moda
for column in ['Escolaridade', 'Tem Filhos', 'Estado Civil', 'Profissão', 'Cor do Cabelo', 'Cor dos Olhos', 'Cor da Pele', 'Tipo de Pele']:
    df_dados[column].fillna(df_dados[column].mode()[0], inplace=True)

# ajusta colunas com duas casas decimais
for column in ['Salário', 'Idade', 'Altura (cm)', 'Peso (kg)']:
    df_dados[column].fillna(round(df_dados[column].mean(), 2), inplace=True)

# Filtrar o dataframe para a região 'Sul'
df_sul = df_dados[df_dados['regiao'] == 'Sul']

# Calcular o desvio padrão da coluna 'Idade'
desvio_padrao_idade = df_sul['Idade'].std()

# Arredondar o resultado para duas casas decimais
desvio_padrao_idade_arredondado = round(desvio_padrao_idade, 2)

# Exibir o desvio padrão e o número de entradas
print(f'O desvio padrão das idades das pessoas que moram na região Sul é: {desvio_padrao_idade_arredondado}')
print(f'Número de entradas (linhas) que entraram no cálculo: {len(df_sul)}')
```

```python
# Filtrar o dataframe para as linhas onde o estado é 'São Paulo' e a pessoa tem filhos
df_sp_filhos = df_dados[(df_dados['estado'] == 'São Paulo') & (df_dados['Tem Filhos'] == 'Sim')]

# Calcular a média de peso
media_peso_sp_filhos = df_sp_filhos['Peso (kg)'].mean()

# Arredondar para duas casas decimais
media_peso_sp_filhos_arredondado = round(media_peso_sp_filhos, 2)

# Exibir a média de pesos calculada
print(f'A média de pesos das pessoas que moram em São Paulo e que possuem filhos é: {media_peso_sp_filhos_arredondado} kg')

# Exibir o número de entradas (linhas) que entraram no cálculo
print(f'Número de entradas (linhas) que entraram no cálculo: {len(df_sp_filhos)}')

df_sp_filhos
```

```python
# Filtrar o dataframe para a profissão 'Professor'
df_professores = df_dados[df_dados['Profissão'] == 'Professor']

# Agrupar por estado e contar ocorrências
contagem_por_estado = df_professores['estado'].value_counts()

# Ordenar em ordem crescente
contagem_por_estado_ordenado = contagem_por_estado.sort_values(ascending=True)

# Imprimir o estado com o menor número de professores e a quantidade
estado_com_menos_professores = contagem_por_estado_ordenado.index[0]
numero_de_professores = contagem_por_estado_ordenado[0]
print(f'O estado com o menor número de professores é: {estado_com_menos_professores}, com {numero_de_professores} professores.')
```
